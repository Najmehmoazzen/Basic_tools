```python
#import librries

import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Create an Erdos_Renyi network
G=nx.erdos_renyi_graph(5,0.5)                       #erdos_renyi_graph(n, p, seed=None, directed=False)


##properties
print("node degree clustering")
for v in nx.nodes(G):
    print(f"{v} {nx.degree(G, v)} {nx.clustering(G, v)}")

#Adjacency matrix as a list
print()                                              # go to next line
print("the adjacency list")
for line in nx.generate_adjlist(G):                  #صفر به چیا وصله، یک به چیا وصله    #balamosalasi (ینی نگفته اگه صفر به 99 وصله، 99 هم به صفر وصله)
    print(line)
#nx.write_adjlist(G, "test.txt")                    # Save adjecency matrix in a txt file


# Get the adjacency matrix
print()                                             # go to next line
print("the adjacency matrix is:")
A = nx.adjacency_matrix(G).todense()                # A is adjecency matrix
print(A)
```

### Another way to create an Erdos Renyi network

```python
def generate_er(n=10, r=0.5):
    """ Function to generate graph networks using 
    Erdos-Renyi Model
    Parameters
    ----------
    n: int
      Number of nodes
    r: float
      probability of the edge
    Returns
    -------
    random ER graph
    """
    return nx.erdos_renyi_graph(n, r)

def plot_er(G, n):
    """ Plots the Erdos-Renyi graph
    Parameters
    ----------
    G: graph
    n: int
      Number of nodes
    Returns
    -------
    None
    """
    # Put the nodes in a circular shape
    pos = nx.circular_layout(G)
    #pos = nx.spring_layout(G, iterations=50, seed=42)
    nx.draw(G, pos, node_color=range(10), node_size=400, cmap=plt.cm.Blues)
    plt.show()
    
# Generate the ER graphs
n=10
p=0.5
G_er = generate_er(n, p)
plot_er(G_er, n)
```

### Another way to find adjacency matrix

```python
import numpy as np

# Set the number of nodes and the probability of connection
n = 10
p = 0.3

# Create the adjacency matrix
adj_matrix = np.zeros((n, n))
for i in range(n):
    for j in range(i+1, n):
        if np.random.rand() < p:
            adj_matrix[i,j] = 1
            adj_matrix[j,i] = 1

print(adj_matrix)
```
